```python
import zipfile
zip_ref=zipfile.ZipFile('/content/dogsvscats.zip','r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
from tensorflow.keras.layers import Activation
from keras.applications.vgg16 import VGG16
```

```python
#generators
train_ds=keras.utils.image_dataset_from_directory(
    directory='/content/train',#train data path
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)

validation_ds=keras.utils.image_dataset_from_directory(
    directory='/content/test', #test data path
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
#normalize
def process(image,label):
  image=tf.cast(image/255. ,tf.float32)
  return image,label
train_ds=train_ds.map(process)
validation_ds=validation_ds.map(process)
```

```python
#CREATE CNN MODEL:
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])  #loss like this as binary classifcation problem
```

```python
history=model.fit(train_ds,epochs=10,validation_data=validation_ds) #run model
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'],color='red',label='train')
plt.plot(history.history['val_accuracy'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
print(history.history.keys())
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'],color='red',label='train')
plt.plot(history.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
#TUNING adding dropout and batch normalization too..
model_improved=Sequential()
model_improved.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3)))
model_improved.add(BatchNormalization())
model_improved.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model_improved.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model_improved.add(BatchNormalization())
model_improved.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model_improved.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model_improved.add(BatchNormalization())
model_improved.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model_improved.add(Flatten())
model_improved.add(Dense(128,activation='relu'))
model_improved.add(Dropout(0.1))
model_improved.add(Dense(64,activation='relu'))
model_improved.add(Dropout(0.1))
model_improved.add(Dense(1,activation='sigmoid'))
```

```python
model_improved.summary()
```

```python
model_improved.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history_improved=model_improved.fit(train_ds,epochs=10,validation_data=validation_ds) #run model
```

```python
print(history_improved.history.keys())
```

```python
plt.plot(history_improved.history['loss'],color='red',label='improved modeltrain loss')
plt.plot(history_improved.history['val_loss'],color='blue',label='improved model validation loss')
plt.legend()
plt.show()
```

```python
plt.plot(history_improved.history['accuracy'],color='red',label='improved model_train_accuracy')
plt.plot(history_improved.history['val_accuracy'],color='blue',label='improved model_validation_accuracy')
plt.legend()
plt.show()
```
